#### Question 1

Let's explore the accuracy of numerical differentiation.

**(a)** Write two functions called twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) that take as arguments arrays of values x and y(x) and return the derivatives of y with respect to x. Make sure that the arrays returned by the function have the same shape as the input arrays.

```python
import numpy as np
def twoPtForwardDiff(x,y):
    """Calculates dy/dx as a forward finite difference"""
    dydxForward = np.zeros(y.shape,float)
    dydxForward[0:-1]= np.diff(y)/np.diff(x)
    dydxForward[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydxForward

def twoPtCenteredDiff(x,y):
    """Calculated dy/dx as a centered finite difference"""
    dydxCenter = np.zeros(y.shape,float)
    dydxCenter[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    dydxCenter[0] = (y[1]-y[0])/(x[1]-x[0])
    dydxCenter[-1] = (y[-1]-y[-2])/(x[-1] -x[-2])
    return dydxCenter
```

**(b)** Use your functions to take the derivative of the function $y = 8x^{4} + 3x$ over the range $0 < x < 10$, with steps of 0.1. How do you know they work correctly? (They should be almost identical to the analytical results).

```python
u = np.arange(0,10.1,0.1)
v = 8*u**4 + 3*u
dydxExact = 32*u**3 + 3
plt.plot(u,twoPtForwardDiff(u,v),color = 'r', 
         label = 'Forward Differentiation')
plt.plot(u,twoPtCenteredDiff(u,v),color = 'g', 
         label = 'Centered Differentiation')
plt.plot(u,dydxExact,color = 'b', 
         label = 'Analytical Differentiation')
plt.legend(loc='upper left')
plt.show
```

Considering that the results are almost on top of the Analytical results, we can tell that the functions worked correctly.

**(c)** What is the accuracy of the results if you use a spacing of 0.01? (Plot the relative error)

```python
u = np.arange(0,10.01,0.01)
v = 8*u**4 + 3*u
dydxExact = 32*u**3 + 3
ForwardError = 100*abs(twoPtForwardDiff(u,v)[1:]-dydxExact[1:])/dydxExact[1:]
CenteredError = 100*abs(twoPtCenteredDiff(u,v)[1:]-dydxExact[1:])/dydxExact[1:]

plt.figure(figsize=[12,8])
plt.semilogy(u[1:], ForwardError, color = 'r',label = 'forward difference')
plt.semilogy(u[1:], CenteredError, color = 'g',label = 'center difference')
plt.ylabel("percent error relative to dydxExact")
plt.legend(loc="upper right")
plt.title("Percent error in multiple differentiation techniques",fontsize=20)
plt.show()
```

The centered difference is always within 0.1% of the correct value, while the forward difference is always within 2% of the correct value.

This method can be used to obtain dydx[2:-2]. **(c)** Write your three differentiating functions to a file called Differentiators.py. Be sure to include meaningful docstrings that describe what the functions do. The docstrings should include enough information for someone to understand the differences between the algorithms. Test that the code works by importing the functions and calling them on $y=8x^{4}+3x$. Plot the results from all three methods.

```python
%%file Differentiators.py
import numpy
def twoPtForwardDiff(x,y):
    """Calculates dy/dx as a forward finite difference.
    The derivative at a point a is the change in f(a) to f(b),
    divided by (b-a), where b > a"""
    dydxForward = numpy.zeros(y.shape,float)
    dydxForward[0:-1]= numpy.diff(y)/numpy.diff(x)
    dydxForward[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydxForward

def twoPtCenteredDiff(x,y):
    """Calculated dy/dx as a centered finite difference. 
    If a < b < c, the derivative of f at b is f(c) - f(a) divided by (c-a)."""
    dydxCenter = numpy.zeros(y.shape,float)
    dydxCenter[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    dydxCenter[0] = (y[1]-y[0])/(x[1]-x[0])
    dydxCenter[-1] = (y[-1]-y[-2])/(x[-1] -x[-2])
    return dydxCenter

def fourPtCenteredDiff(x,y):
    """Calculated dy/dx as a centered finite difference with four points of reference. If a < b < c < d < h, the derivative of f at b is calculated as f(a) - 8f(b) + 8f(d) - f(h), divided by 3(h-a)."""
    dydxcenter = numpy.zeros(y.shape,float)
    dydxcenter[2:-2] = (y[0:-4] - 8*y[1:-3] + 8*y[3:-1] - y[4:])/(12*(x[2:-2] - x[1:-3]))
    dydxcenter[0] = (y[1]-y[0])/(x[1]-x[0])
    dydxcenter[-1] = (y[-1]-y[-2])/(x[-1] - x[-2])
    dydxcenter[1] = (y[1] - y[3])/(x[1]-x[3])
    dydxcenter[-2] = (y[-1]-y[-3])/(x[-1] - x[-3])
    return dydxcenter
```
