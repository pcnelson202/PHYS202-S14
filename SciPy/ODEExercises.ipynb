### Question 1

For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE tour.


**(a)** See what happens to the period of the pendulum as you change the initial angle of release. Plot $\theta$ vs. $t$ for 5 different initial angles spanning 0 to $\pi$. Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from $\pi$/4 to 3$\pi$/4? What happens to the trajectory when the initial angle is 0 and $\pi$?

```python
def deriv(theta_current,t):
    g = 9.8
    ell = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta,domega])


t = np.arange(0.,10.01,0.01)
th0 = 0.
omega0 = 0.
theta_init = np.array([th0,omega0])
init_0 = odeint(deriv,[0.,0.],t)
init_1 = odeint(deriv,[np.pi/4,0.],t)
init_2 = odeint(deriv,[np.pi/2,0.],t)
init_3 = odeint(deriv,[3*np.pi/4,0.],t)
init_4 = odeint(deriv,[np.pi,0.],t)

plt.plot(t,init_0[:,0],'r--',label="")
plt.plot(t,init_1[:,0],'b--',label="$\theta_0 = \pi/4$")
plt.plot(t,init_2[:,0],'g--',label="Numerical solution")
plt.plot(t,init_3[:,0],label="Numerical solution")
plt.plot(t,init_4[:,0],label="Numerical solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.grid()
plt.legend()
plt.show()
```
