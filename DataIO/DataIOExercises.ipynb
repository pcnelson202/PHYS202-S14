**1.** Here is some code that creates a comma-delimitedd file of numbers with random precision, leading spaces, and formatting:

```python
f = open('messy_data.dat', 'w')
import random
for i in range(100):
    for j in range(5):
        f.write(' ' * random.randint(0,6))
        f.write('%0*.*g' % (random.randint(8,12),
                            random.randint(5,10),
                            100 * random.random()))
        if j != 4:
            f.write(',')
    f.write('\n')
f.close()
```
