# Ordinary Differential Equations

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

### The Simple Pendulum as an example

Solving for the motion of a simple pendulum involves a second order differential equation that cannot be solved analytically. If you use Newton's second law to write the sum of the forces on the mass, you get $\sum\vec{F} = \vec{T} + \vec{w} = m\vec{a}$

This leads to the differential equation:

\begin{equation}
\frac{d^{2}\theta}{dt^2} = -\frac{g}{\ell}sin(\theta)
\end{equation}

Because of the $sin\theta$ on the right side, this has no analytic solution. So we can either approximate it, or solve it numerically.

### Approximate solution

Using the small angle approximation, we get:

\begin{equation}
\theta(t) = \theta_{0}cos\left(\sqrt\frac{g}{\ell}t\right)
\end{equation}

```python
g = 9.8
ell = 0.5
th0 = 20.*np.pi/180

t = np.arange(0.,10.+0.01,0.01)
theta_app = lambda t: th0*np.cos(np.sqrt(g/ell)*t)

plt.plot(t,theta_app(t), label='Approximate solution')
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize = 20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

### Numerical Solution

We'll define $\omega$ so that:

\begin{equation}
\frac{d^{2}\theta}{dt^2} = \frac{d\omega}{dt} = -\frac{g}{\ell}sin\theta
\end{equation}

Now we can use scipy's ODE solver, called odeint, to solve this system of equations. We need to create a function that returns the first derivative of each element in the array. Here's what it looks like.

```python
def deriv(theta_current,t):
    g = 9.8
    ell = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta,domega])
```

```python
from scipy.integrate import odeint
```

```python
t = np.arange(0.,10.01,0.01)
omega0 = 0.
theta_init = np.array([th0,omega0])

theta = odeint(deriv,theta_init,t)
```

```python
print theta.shape
```

```python
plt.plot(t,theta_app(t),'b-',label="Approximate solution")
plt.plot(t,theta[:,0],'r--',label="Numerical solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

Much more significant deviations appear for large angles. To get a good estimate of the motion of a simple pendulum over a wide range of motion, numerical solutions are the way to go. The nonlinear behavior will lead to some surprising physical results.

### Breaking into the black box

In order to be able to solve for the motion of the simple pendulum using odeint, we had to convert the single second-order ODE into two first order ODEs. Let's peer inside scipy's black box to understand why.

#### Euler's Method

The most basic numerical algorithm for computing the solutions to ODEs was invented by Leonhard Euler in the 18th century - way before microprocessors.

#### Freefall as an example

\begin{equation}
a(t) = \frac{d^2y}{dt^2}
\end{equation}

The Euler Method involves differentiation and is based on a tangent line approximation.

```python
def EulerFreefall():
    tmax = 10
    dt = 0.1
    a = -9.81
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y=np.zeros(nsteps)
    t[0]=0
    v[0] = 0
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1]=v[i] + dt*a
        y[i+1]=y[i]+dt*v[i]
    return t,v,y
te,ve,ye = EulerFreefall()
```

```python
plt.plot(te,ye,"r-",label="y(t)")
plt.plot(te,ve,"b-",label="v(t)")
plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),v(t)",fontsize=20)
plt.grid()
plt.legend(loc="best", fontsize=20)
plt.show()
```

#### Exploring the parameter space

```python
def freefall(tmax=10,dt=0.1,t0=0.,v0=46.,y0=0.):
    a =-9.81
    t = np.zeros(int(tmax/dt))
    v = np.zeros(int(tmax/dt))
    y = np.zeros(int(tmax/dt))
    t[0] = t0
    v[0] = v0
    y[0] = y0
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt*a
        y[i+1] = y[i] + dt*v[i]
    tmax = -v0/a
    ymax = y0 + v0*tmax + 0.5*a*tmax**2
    percentDiff = 100.*np.abs(y.max()-ymax)/ymax
    print "ymax percentDiff = %.2f"%percentDiff
    plt.plot(t,y,"r-",label="y(t)")
    plt.plot(t,v,"b-",label="v(t)")
    plt.xlabel("time (s)",fontsize = 20)
    plt.ylabel("y(t),v(t)",fontsize=20)
    plt.xlim(0,20.)
    plt.ylim(-100.,600.)
    plt.grid()
    plt.legend(loc = "upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
```

#### Euler-Cromer Method

Varying the timesteps let's us improve the accuracy of our result, but to do an even better job, we could try modifying the method for calculating the new positions and velocities. Cromer added the idea of calculating the new position using just the calculated new velocity at the end of the interval to Euler's method, creating what is known as the Euler-Cromer Method.

```python
def EulerCromerFreefall():
    tmax = 10
    dt = 0.1
    a = -9.81
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y=np.zeros(nsteps)
    t[0]=0
    v[0] = 0
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1]=v[i] + dt*a
        y[i+1]=y[i]+dt*v[i+1]
    return t,v,y
```

```python
tec,vec,yec = EulerCromerFreefall()
```

```python
plt.plot(tec,yec,"r-",label="y(t)")
plt.plot(tec,vec,"b-",label="v(t)")
plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),v(t)",fontsize=20)
plt.grid()
plt.title("Euler-Cromer")
plt.legend(loc = "best",fontsize=20)
plt.show()
```

### Midpoint Method

\begin{equation}
y_{i+1} = y_i + \frac{1}{2}(v_i + v_{i+1})\Delta t
\end{equation}

### The Runge-Kutta Family of Methods

The midpoint method and Euler's methods are the order 1 and 2 parts of the Runge-Kutta Family of methods.

### What's inside scipy.integrate.odeint?

A Fortran algorithm called LSODE(Livermore Solver for ODE's)

### Coupled differential equations

\begin{equation}
\frac{dZ}{dt} = \alpha Z - \beta ZL
\end{equation}

\begin{equation}
\frac{dL}{dt} = \epsilon\beta ZL - \gamma L
\end{equation}

* $\alpha$ = 0.7 per year
* $\gamma$ = 0.5 per year
* $\beta$ = 0.007 per lion per year
* $\epsilon$ = 0.1 lions per zebra

```python
def PPderiv(y,t):
    alpha = 0.7
    gamma = 0.5
    beta=0.007
    eps=0.1
    Zi=y[0]
    Li=y[1]
    dZdt = alpha*Zi - beta*Zi*Li
    dLdt = eps*beta*Zi*Li - gamma*Li
    return np.array([dZdt,dLdt])
```

```python
t = np.arange(0.,50.1,0.1)
Z0 = 200
L0 = 50
result = odeint(PPderiv,np.array([Z0,L0]),t)
Z = result[:,0]
L = result[:,1]
plt.plot(t,Z,label="Zebra population")
plt.plot(t,L,label="Lion population")
plt.xlabel("time (years)")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
res1 = odeint(PPderiv,np.array([100,50]),t)
res2 = odeint(PPderiv,np.array([300,50]),t)
plt.plot(res1[:,0],res1[:,1],label="Z$_0$ = 100, L$_0$ = 50")
plt.plot(Z,L,label="Z$_0$ = 200, L$_0$ = 50")
plt.plot(res2[:,0],res2[:,1],label="Z$_0$ = 300, L$_0$ = 50")
plt.xlabel("Zebras",fontsize=20)
plt.ylabel("Lions",fontsize=20)
plt.legend()
plt.show()
```
